```python
X_train, X_test, y_train, y_test = train_test_split(data.iloc[:,0:8], data.iloc[:,8:10], test_size=0.20, random_state=42)
scaler = StandardScaler()
scaler.fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
rgrs = MLPRegressor(hidden_layer_sizes=(10,16,16,16,10,5,3))
rgrs.fit(X_train, y_train)
accuracy = rgrs.score(X_test, y_test)
print "Accuracy is %.2f%%" %(accuracy*100) 
```
